### Creating basic tensor data 

```python
import tensorflow as tf
print(tf.__version__)
import numpy as np
```

```python
# Create tensors as tf.constant()
scalar = tf.constant(7)
scalar
```

```python
# Check number of dimensions of a tensor
scalar.ndim
```

```python
# Create vector

vector = tf.constant([10, 10])
print(vector)
print(vector.ndim)
```

```python
# Create matix (has more than one dim)

matrix = tf.constant([[10, 7], [7, 10]])

print(matrix)
print(matrix.ndim)
```

```python
# Create a tensor

tensor = tf.constant([[[1, 2, 3], 
                       [1,2,3]],
                       [[1,2,3], 
                        [3,4,5]], 
                      [[3,4,5], 
                       [3,4,5]]])

print(f"{tensor}, number of dims {tensor.ndim}")
```

```python
# Creating tensor with tf.Variable()
changable_tensor = tf.Variable([10, 7])
unchangable_tensor = tf.constant([10, 7])

changable_tensor, unchangable_tensor
```

```python
# Changing elements
changable_tensor[0].assign(7)
changable_tensor
```

### Creating random tensors
Random tensors are tensors of some abitrary size which contain random numbers.

e.g. weights are random initialized numbers (well not exactly normal distribution/ glorify etc...)

from_seed is giving extra x value so everything using the 42 if going to be the same


```python
random_1 = tf.random.Generator.from_seed(42)
random_1 = random_1.normal(shape=(3,2))
random_1
```

```python
random_2 = tf.random.Generator.from_seed(42)
random_2 = random_2.normal(shape=(3,2))
random_2
```

```python
tf.reduce_mean(random_1), tf.reduce_mean(random_2)
```

### Shuffle order of elements in tensor/

```python
not_shuffled = tf.constant([[10,7],
                            [3, 4],
                            [2, 3]])
not_shuffled
```

```python
shuffled = tf.random.shuffle(not_shuffled, seed=42)
shuffled
```

```python
tf.random.set_seed(7)
tf.random.shuffle(not_shuffled, seed=7)
```

When you initialize a global seed (tf.random.set_seed(42)) and you pass that to the operation seed in the function, it uses the global state of X param.

Thats how if we want ouf shuffled tensors to be in the same order, we have ti use global seed

### Getting infos from Tensors

* Shape - The length of each of dimensions of tensor - .shape
* Rank - Nuber of dims - .ndim
* Axis of dimesnion - a particular dimension of tensor - tensor[0], tensor[:, 1]
* Size - Total number of items in tensor - tf.size(tensor)

```python
# Create a rank 4 tensor (4 dimensions)

rank_4_tensor = tf.zeros(shape=(2, 3, 4, 5))
rank_4_tensor
```

```python
rank_4_tensor.shape, rank_4_tensor.ndim, tf.size(rank_4_tensor)
```

```python
2*3*4*5
```

```python
# Get various attributes of our tensor
print("Datatype of every element:", rank_4_tensor.dtype)
print("Number of dims:", rank_4_tensor.ndim)
print("Shape of tensor:", rank_4_tensor.shape)
print("Elements along the 0 axis:", rank_4_tensor.shape[0])
print("Elements along the last axis:", rank_4_tensor.shape[-1])
print("Total number of elements in our tensor:", tf.size(rank_4_tensor).numpy())
```

#### Index tensors

Tensors can be indexed like Python lists.

```python
# Get the first 2 elements of each dim

rank_4_tensor[:, :2, :2, :2]
```

```python
# Get the first element from each dim from each index except final one

rank_4_tensor[:1, :1, :, :]
```

```python
rank_2_tensor = tf.constant([[2,4], [15,16]])

# Get last item of each of our 2 rank tensor

rank_2_tensor[:, -1]
```

```python
# ADD in extra dim from rank 2 to rank 3 tensor
## ... means every axis before to be the same
rank_3_tensor = rank_2_tensor[..., tf.newaxis]
rank_3_tensor
```

```python
tf.expand_dims(rank_2_tensor, axis=-1) # -1 because of the last axis
```

**Matix multiplication**

In ML this is one of the most common tensor operations

There are two rules our tensors need to fultil if we're going to matix multiply them: 

1. The inner dims must match e.g (3 x 4) * (4 x 5)
2. The resulting matrix has the shape of the inner dimensions

```python
tf.matmul(tensor, tensor)
```

```python
# in python me have different operator "@"
tensor @ tensor
```

```python
tensor * tensor[..., tf.newaxis]
```

### Aggregating tensors

Aggregation tensors = condensing them from mltiple values down to a smaller amount of values

min, max, mean, sum of tensor

```python
# Get the absolute values
D = tf.constant([-7, -10])
D, tf.abs(D)
```

```python
E = tf.Variable(np.random.randint(0,100, size = (3,2)), shape=(3, 2))
E
```

```python
print("Min:",tf.reduce_min(E).numpy()) 
print("Max:", tf.reduce_max(E).numpy())
print("Mean:", tf.reduce_mean(E).numpy()) 
print("Sum:", tf.reduce_sum(E).numpy()) 
```

```python
tf.math.reduce_variance(tf.cast(E, dtype=tf.float16))
```

```python
tf.math.reduce_std(tf.cast(E, dtype=tf.float16))
```
